### Clustering Assignment with Pyspark K-means

```python
# Import libraries 
from numpy import array
from math import sqrt

# Import pyspark machine learning library (mllib) packages 
from pyspark.mllib.clustering import KMeans, KMeansModel
```

```python
# Import libraries and packages - to read in dataset and 
import pandas as pd
from pyspark import SparkContext

# Uncomment next staement for repeat statement runs of sc
# sc.stop()
```

```python
# set up a client to the Spark environment
sc = SparkContext()
```

```python
# Read in two of the 'Walking Activity' datasets, and then parse the data

import csv
    
data1 = pd.read_csv("Data/1.csv",header=None, names=['timestep', 'xaccel', 'yaccel', 'zaccel']) # no header in file

data2 = pd.read_csv("Data/2.csv",header=None, names=['timestep', 'xaccel', 'yaccel', 'zaccel'])

type(data1)
#display(data1)
data1.head()
data1.columns

df1=data1[['xaccel', 'yaccel', 'zaccel']]
df1.columns
len(df1)


df2=data2[['xaccel', 'yaccel', 'zaccel']]
df2.columns
len(df2)

df=pd.concat([df1, df2])
len(df)
#type(df)
```

```python
# This step is used to convert pandas DataFrame -> Spark DataFrame -> Spark RDD

from pyspark.sql import SQLContext

sqlctx = SQLContext(sc)

#Â convert your pandas dataframe to a spark dataframe
sdf = sqlctx.createDataFrame(df)

# register the table to use it across interpreters
sdf.registerTempTable("sdf")

# get the underlying RDD
rdd = sdf.rdd

print("The sdf is of type %s and the rdd is of type %s" % (type(sdf), type(rdd)) )

```

```python
# View the RDD dataset
for i in rdd.collect():
   print(i)
```

```python
# Save dataframe to file to be read by spark instance -> pyspark RDD
df[:20].to_csv('clusterdata.txt', columns = ['xaccel', 'yaccel', 'zaccel'], header=False, index=False)

data = sc.textFile("clusterdata.txt")

# Parse the pyspark RDD 
parseddata = data.map(lambda line: array([float(x) for x in line.split(',')]))

print('The number of observations', parseddata.count())
```

```python
for x in data.collect():
    print(x)
    
    
```

```python
# Build the model (cluster the data)
clusters = KMeans.train(parseddata,k=4, maxIterations=10, initializationMode="random")
```

```python
# define procedure to evaluate clustering by computing Within Set Sum of Squared Errors

def error(point):
    center = clusters.centers[clusters.predict(point)]
    return sqrt(sum([x**2 for x in (point - center)]))
```

```python
# Evaluate clustering by computing Within Set Sum of Squared Errors
WSSSE = parseddata.map(lambda point: error(point)).reduce(lambda x, y: x + y)
print("Within Set Sum of Squared Error = " + str(WSSSE))
```

```python
print("Final centers: " + str(clusters.clusterCenters))  
print('\n')
print("Total Cost: " + str(clusters.computeCost(parseddata)))  
```
